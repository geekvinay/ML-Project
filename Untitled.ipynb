```python
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np

print("Fetching all the images")

# Storing all the face data in the 
facesArr = []

counter = 0

label = glob.glob("./dataset/train/*/*.jpg")[0]

for imgAddress in glob.glob("./dataset/train/*/*.jpg"):
    
    print(imgAddress)
    temp = imgAddress.split("\\")[1]
    
    if(temp==label):
        flag = 0
    else:
        flag = 1
    
    label = temp
    
    if(flag==1): 
        counter = 0
    
    # Loading the images in grayscale
    img = cv2.imread(imgAddress)
    tempArr = []
    
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    # Converting those images from BGR scale to grayscale
    test_image_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
    face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_alt.xml')
    faces=face_cascade.detectMultiScale(test_image_gray,scaleFactor=1.05,minNeighbors=9)

    print(faces,len(faces))
    
    # Ignore cases where the  
    if(len(faces)==0): 
        continue
        
    biggestArea = max(faces[:,2]) * max(faces[:,3])
    
    for x,y,w,h in faces:
        imageArea = w*h
        
        print(imageArea,biggestArea)
        
        if (biggestArea == imageArea):
            counter += 1
            
            # print(f"The highest face detected is {x} {y} {w} {h}")
            # img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
            # plt.imshow(img)
            # plt.show()

            #display the cropped image
            crop_img=img[y:y+h,x:x+w]
            crop_img = cv2.resize(crop_img,(512,512))
            
            crop_img = cv2.cvtColor(crop_img,cv2.COLOR_BGR2GRAY)
            
            tempArr.append(crop_img)
            
            # print(f"\n\nClassLabel ==> {label}Counter ==> {counter}")
            cv2.imwrite(("./dataset/test-worked/%s-%d.jpg" % (label,counter)),crop_img)
            # plt.imshow(crop_img)
            # plt.show()
    tempArr = np.array(tempArr)
    facesArr.append(list([label,tempArr]))

print(facesArr)
facesArr = np.array(facesArr)
```
